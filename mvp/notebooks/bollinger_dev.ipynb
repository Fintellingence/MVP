```python
import mvp
import pandas as pd
import numpy as np
import datetime as dt
DB_PATH = '/home/naga/FintelligenceData/MetaTrader_M1.db'
```

```python
db_parser = mvp.builder.Yahoo(DB_PATH)
```

```python
symbols = db_parser.get_db_symbols()
```

```python
symbols
```

```python
abev_raw = mvp.rawdata.RawData('ABEV3',DB_PATH)
```

```python
abev = mvp.curated.CuratedData(abev_raw,{},daily=True)
```

```python
returns = abev.frac_diff(1,0.001,improve=True)
```

```python
returns
```

```python
returns.plot.hist(bins = 70)
```
